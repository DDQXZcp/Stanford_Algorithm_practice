## merge & count

```python
def merge(left, right):
    res = []
    while left and right:
        if left[0] < right[0]:
            res.append(left.pop(0))
        else:
            res.append(right.pop(0))
    res = res + left + right
    return res
```

```python
def merge_and_cnt(nleft, nright): # assume: input numpy array, output numpy array
    left = nleft.tolist()
    right = nright.tolist()
    res = []
    i = 0
    cnt = 0
    size = len(left)
    while left and right:
        if left[0] < right[0]:
            res.append(left.pop(0))
            i=i+1
        else:
            res.append(right.pop(0))
            cnt += size-i
    res = res + left + right
    nres = np.array(res)
    return nres,cnt
```

## internal inversion inside an array

```python
def cnt_inv(A):
    inv_cnt=0
    size = A.size
    for i in range(0,size):
        for j in range(i+1,size):
            if(A[i]>A[j]):
                inv_cnt = inv_cnt+1
        print(i,': ',inv_cnt)
    return inv_cnt
```
